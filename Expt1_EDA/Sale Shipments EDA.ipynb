```python
import pandas as pd
df = pd.read_csv('/content/Sale Shipments Summary Statistics.csv')

df.head()
```

```python
df.info()
```

```python
# Calculating the number of null values in each column

total_null_values = df.isnull().sum().sort_values(ascending=False)
total_null_values
```

```python
# Getting first 5 rows of the 'Number of employees' column

df['Number of employees (EMP)'].head()
```

```python
# Get the distributions of each value for 'Meaning of Tax status code (TAXSTAT_LABEL)' by using value_counts().

df['Meaning of Tax status code (TAXSTAT_LABEL)'].value_counts()
```

```python
# Getting details of the Annual Payroll column

df['Annual payroll ($1,000) (PAYANN)'].describe()
```

```python
# Getting max value of the 'Annual Payroll'

df['Annual payroll ($1,000) (PAYANN)'].max()
```

```python
# A subset of our data having tax status code as 'Establishments subject to federal income tax'

df[df['Meaning of Tax status code (TAXSTAT_LABEL)'] == 'Establishments subject to federal income tax'].head()
```

```python
# Getting all rows for which 'Annual payroll ($1,000) (PAYANN)' < 50000000

df['Low_Annual_Payroll'] = df['Annual payroll ($1,000) (PAYANN)'] < 50000000
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
sns.set()
sns.set_context('talk')
```

### Univariate Plotting

```python
sns.displot(df['Number of employees (EMP)'], kde=True)
```

```python
sns.displot(df['Number of employees (EMP)'], bins=10, kde=True)
```

### Bivariate Plotting

```python
sns.jointplot(x=df['Sales, value of shipments, or revenue ($1,000) (RCPTOT)'],y=df['Annual payroll ($1,000) (PAYANN)'])
```

```python
sns.jointplot(x=df['Sales, value of shipments, or revenue ($1,000) (RCPTOT)'],y=df['Annual payroll ($1,000) (PAYANN)'], kind='reg')
```

```python
sns.jointplot(x=df['Sales, value of shipments, or revenue ($1,000) (RCPTOT)'],y=df['Annual payroll ($1,000) (PAYANN)'], kind='kde', shade=True)
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(),cbar=True,annot=True, cmap='Blues')
```
